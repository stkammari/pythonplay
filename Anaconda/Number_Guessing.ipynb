```python

import random
simple_number = random.randint(1, 10)
print("Welocme to number guessing game!")
print("I am thinking of a number between 1 and 10")

# input to user for Guessing number 
guess_number = int(input("Guess a number: ") )

# check if the guess is correct 
if guess_number == simple_number:
    print(f"Yes, your guess is correct!")
else:
    print(f"The number was {simple_number}. Better luck next time!")
```
